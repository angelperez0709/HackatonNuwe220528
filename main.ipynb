```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
def getPrepareData(url_csv):
  dataset = pd.read_csv(url_csv)
  dataset = dataset.sort_values("Date")
  dataset = dataset.drop("Date",axis=1)
  #Delete rows with nan
  dataset = dataset.dropna()
  #Normalization data
  sc = MinMaxScaler(feature_range=(0,1))
  normal_dataset = sc.fit_transform(dataset)
  return normal_dataset
```

```python
#Get data for training
normal_dataset = getPrepareData('https://raw.githubusercontent.com/angelperez0709/archivos/main/train.csv')
```

```python
def transformData(data):
  days = 3
  x_train = []

  for i in range(0,len(data) - days):
    if(i > 2):
      mean_open = data[i:i+days,0].mean()
      mean_high = data[i:i+days,1].mean()
      mean_low = data[i:i+days,2].mean()
      mean_close = data[i:i+days,3].mean()
      mean_adj = data[i:i+days,4].mean()
      mean_volume = data[i:i+days,5].mean()
      x_train.append([mean_open,mean_low,mean_close,mean_adj,mean_volume])
    else:
      mean_open = data[i:i+days,0].mean()
      mean_high = data[i:i+days,1].mean()
      mean_low = data[i:i+days,2].mean()
      mean_close = data[i:i+days,3].mean()
      mean_adj = data[i:i+days,4].mean()
      mean_volume = data[i:i+days,5].mean()
      x_train.append([mean_open,mean_low,mean_close,mean_adj,mean_volume])
      mean_open = data[i:i+days,0].mean()
      mean_high = data[i:i+days,1].mean()
      mean_low = data[i:i+days,2].mean()
      mean_close = data[i:i+days,3].mean()
      mean_adj = data[i:i+days,4].mean()
      mean_volume = data[i:i+days,5].mean()
      x_train.append([mean_open,mean_low,mean_close,mean_adj,mean_volume])

  x_train = np.array(x_train)
  return x_train

def transformDataY(data):
  y_train = []
  days = 3
  for i in range(0,len(data) - days):
    if(i > 2):
      y_train.append(data[i+days,6])
    else:
      y_train.append(data[i+days,6])
      y_train.append(data[i+days,6])
  y_train = np.array(y_train)
  return y_train
```

```python
x_train = transformData(normal_dataset)
y_train = transformDataY(normal_dataset)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
grid = {'n_estimators': [100, 200, 500],
        'max_depth': [20, 30],
         'max_features': ['auto','sqrt'],
         'min_samples_split': [2],
         'min_samples_leaf': [4]}
model = RandomForestClassifier()
gs_model = GridSearchCV(estimator=model,param_grid=grid,cv=5,verbose=2)
gs_model.fit(x_train,y_train)
```

```python
x_test = getPrepareData('https://raw.githubusercontent.com/angelperez0709/archivos/main/test_x.csv')
```

```python
x_test = transformData(x_test)
```

```python
y_preds = gs_model.predict(x_test)
```
